```python
sns.barplot(x='Gender', y='Age',data = df)
```

```python
sns.countplot(data=df,x='Age', hue='Gender')
sns.set(rc={'figure.figsize':(15,25)})
```

```python
sns.barplot(x='Product Category', y='Price per Unit',data = df)
sns.set(rc={'figure.figsize':(5,25)})
```

```python
sns.scatterplot(x ='Quantity',y ='Total Amount',data=df,hue='Product Category')
sns.set(rc={'figure.figsize':(5,15)})
```

```python
r=df[df["Total Amount"]>100]
r1=r.groupby(["Product Category"])["Total Amount"].mean().sort_values(ascending=True)
r1[:3]
```

```python
plt.barh(r1.index,r1)
plt.title("Product Category - Total Amount")
plt.xlabel("Total Amount")
plt.ylabel("Product Category")
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df,x='Price per Unit')
for bars in ax.containers:
    ax.bar_label(bars)
```
